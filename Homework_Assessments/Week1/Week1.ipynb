```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def transform_to_rgb(img): # Created my own cv2.cvtColor(img,COLOR_BGR2RGB) function to transform the images from bgr to rgb
  new_img = img[:, :, ::-1]
  return new_img
```

```python
img  = cv2.imread("/content/redlogo.jpg")
print(img.shape[0],img.shape[1])
```

```python
transpose = np.zeros((img.shape[1], img.shape[0], img.shape[2]), dtype=np.uint8)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    transpose[j][i] = img[i][j]
cv2.imwrite("transpose_img.jpg",transpose)
transpose_rgb = transform_to_rgb(transpose)
transpose_rgb
```

```python
rotate_90_img = np.zeros_like(transpose)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    rotate_90_img[j][img.shape[0]-i-1] = img[i][j]
cv2.imwrite("rotate_90_img.jpg",rotate_90_img)
rotate_90_img_rgb = transform_to_rgb(rotate_90_img)
rotate_90_img_rgb
```

```python
grayscale_img = np.zeros((img.shape[0],img.shape[1]))
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    grayscale_img[i][j] = np.mean(img[i][j])
cv2.imwrite("grayscale_img.jpg",grayscale_img)
plt.imshow(grayscale_img,cmap = 'gray')
```

```python
flip_img = np.zeros_like(img)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    flip_img[i][j] = img[i][img.shape[1]-j-1]
cv2.imwrite("flip_img.jpg",flip_img)
flip_rgb = transform_to_rgb(flip_img)
flip_rgb
```

```python
green_img = np.ones_like(img)*255

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
      if not (img[i, j] == [255, 255, 255]).all():
          green_img[i, j] = [0, img[i][j][1], 0]
cv2.imwrite('green_logo.jpg', green_img)
green_img_rgb = transform_to_rgb(green_img)
green_img_rgb
```

```python
grid_img = np.zeros((img.shape[0]*2,img.shape[1]*2,3),dtype = np.uint8)
x = img.shape[0]
y = img.shape[1]
k = 0
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    grid_img[i][j] = img[i][j]
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    grid_img[i+x][j] = img[i][j]
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    grid_img[i][j+y] = img[i][j]
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    grid_img[i+x][j+y] = img[i][j]
cv2.imwrite('grid_img.jpg', grid_img)
grid_rgb = transform_to_rgb(grid_img)
grid_rgb
```

```python
binary_img = np.zeros((x, y), dtype=np.uint8)
for i in range(x):
    for j in range(y):
        if(int(np.mean(img[i][j])<127)):
          binary_img[i][j] = 255
        else:
          binary_img[i][j] = 0

cv2.imwrite('binary_logo.jpg', binary_img)
plt.imshow(binary_img,cmap = 'gray')
```

```python
trishul_removed = img.copy()

center_x, center_y = x//2+8, y//2-10 # Manually adjusted the centre to erase the trishul
radius = 44

bg = [255, 255, 255]

for i in range(img.shape[0]):
    for j in range(img.shape[1]):

        dist = np.sqrt((j - center_x) ** 2 + (i - center_y) ** 2)
        if dist <= radius:
            trishul_removed[i, j] = bg

cv2.imwrite('trishul_removed.jpg', trishul_removed)
trishul_removed_rgb = transform_to_rgb(trishul_removed)
trishul_removed_rgb
```
